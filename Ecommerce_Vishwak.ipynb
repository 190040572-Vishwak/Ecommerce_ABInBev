```python
df4 = df1.merge(df2,on=["Order ID"])
df4
```

```python
#Profit Based on the Order Id
df11=df4.groupby(['Order ID'])['Profit'].sum()
pd.DataFrame(df11)
```

```python
#Profit Based on the State
df12=df4.groupby(['State'])['Profit'].sum()
pd.DataFrame(df12)
```

```python
#Profit Based on the Category
df13=df4.groupby(['Category'])['Profit'].sum()
pd.DataFrame(df13)
```

```python
#Profit Based on the Sub-Category
df14=df4.groupby(['Sub-Category'])['Profit'].sum()
pd.DataFrame(df14)
```
